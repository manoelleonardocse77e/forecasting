```python
import pandas as pd
import yfinance as yf
from datetime import datetime
from datetime import timedelta
import plotly.graph_objects as go
from prophet import Prophet
from prophet.plot import plot_plotly, plot_comments_plotly
import warnings
warnings.filterwarnings("ignore")
pd.options.display.float_format = "$(:,.2f)".format
```

```python
!pip install prophet
```

```python
hoje = datetime.today().strftime("%Y-%m-%d")
inicio = '2016-01-01'
data_etherium = yf.download("ETH-USD", inicio, hoje)
data_etherium.tail()
```

```python
data_etherium.reset_index(inplace=True)
```

```python
data_etherium
```

```python
data = data_etherium[["Date", 'Adj Close']]
data.rename(columns = {"Date" : 'ds', 'Adj Close' : "y"}, inplace=True)
```

```python
data
```

```python
!pip install plotly
```

```python
from plotly.graph_objects import Figure, Scatter
fig = go.Figure()
fig.add_trace(go.Scatter(x=data['ds'], y=data['y']))
```

```python
model = Prophet( seasonality_mode='multiplicative')
model.fit(data)
```

```python
data_futuro = model.make_future_dataframe(periods=60)
data_futuro.tail()
```

```python
prever = model.predict(data_futuro)
prever.tail()
```

```python
plot_plotly(model, prever)
```

```python
from prophet.plot import plot_components_plotly
```

```python
plot_components_plotly(model, prever)
```
